```python
import re
def clean_text(text):
    # ç§»é™¤è¡¨æƒ…ç¬¦å·
    text = re.sub(r'[^\w\s,]', '', text)
    
    # ç§»é™¤URL
    text = re.sub(r'http\S+', '', text)
     
    return text
```

```python
# translate all the description into English
from googletrans import Translator
def translate_description_intoeng(description):
    translator = Translator()
    #
    description = clean_text(description)
    # Detect text language
    detected = translator.detect(description)
    
    # If the detected language is not English, translate it
    if detected.lang != 'en':
        translated = translator.translate(description, dest='en')
        return translated.text
    else:
        # If the text is already in English, return it as is
        return description

```
